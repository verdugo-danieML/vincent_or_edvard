```python
def search_images(term, max_images=1):
    print(f"Searching for '{term}'")
    return L(ddg_images(term, max_results=max_images)).itemgot('image')
```

```python
urls = search_images("Van gogh's chair - by Vincent van Gogh", max_images=1)
urls[0]
```

```python
from fastdownload import download_url
dest = 'vincent_van_gogh.jpg'
download_url(urls[0], dest, show_progress=False)

from fastai.vision.all import *
im = Image.open(dest)
im.to_thumb(256,256)
```

```python
download_url(search_images('The Scream 1893 - by Edvard Munch', max_images=1)[0], 'edvard_munch.jpg', show_progress=False)
Image.open('edvard_munch.jpg').to_thumb(256,256)
```

```python
searches = {'vincent_van_gogh' : vincent_van_gogh,
            'edvard_munch' : edvard_munch}

list(searches.keys())
```

```python
path = Path('vincent_or_edvard')
from time import sleep

for key, values in searches.items():
    dest = (path/key)
    dest.mkdir(exist_ok=True, parents=True)
    for img_link in values:
        download_images(dest, urls=search_images(img_link))
    resize_images(path/key, max_size=400, dest=path/key)
```

## Step 2: Train our model

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink)
```

```python
len(failed)
```

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms()).dataloaders(path, bs=32)
```

```python
dls.show_batch(max_n=6)
```

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(10)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5,nrows=5)
```

```python
#|export

learn_inf = load_learner('model.pkl')
```

```python
learn_inf.predict('vincent_van_gogh.jpg')
```

```python
learn_inf.dls.vocab
```

```python
#|export

categories = ('edvard_munch', 'vincent_van_gogh')

def classify_images(img):
    pred, idx,probs = learn_inf.predict(img)
    return dict(zip(categories, map(float,probs)))
```

```python
classify_images('vincent_van_gogh.jpg')
```

```python
#|export

image = gr.inputs.Image(shape=(256,256))
label = gr.outputs.Label()
examples = ['vincent_van_gogh.jpg', 'edvard_munch.jpg']

intf = gr.Interface(fn= classify_images, inputs=image, outputs=label, examples=examples)
intf.launch(inline=False)
```
